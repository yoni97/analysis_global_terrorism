```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../csv_files/globalterrorismdb.csv', encoding='latin1')
df = df.drop_duplicates()
df.head()
```

```python
df.rename(columns={'iyear':'Year', 'imonth':'Month', 'iday':'Day', 'country_txt':'Country', 'provstate':'state',
                     'region_txt':'Region', 'attacktype1_txt':'AttackType', 'target1':'Target', 'nkill':'Killed',
                     'nwound':'Wounded', 'summary':'Summary', 'gname':'Group', 'targtype1_txt':'Target_type',
                     'weaptype1_txt':'Weapon_type', 'nperps': 'nperps'
                    },inplace=True)
df
```

```python
df = df[['eventid','Year','Month','Day','Country','Region','state','city','latitude','longitude','AttackType','Killed','Wounded',
            'Target','Summary','Group','Target_type','Weapon_type', 'nperps', 'success']]
df
```

```python
data = pd.read_csv('../csv_files/RAND_Database_of_Worldwide_Terrorism_Incidents.csv', encoding='latin1', low_memory=False)
data
```

```python
data['Date'] = pd.to_datetime(data['Date'], format='%d-%b-%y', errors='coerce')
# print(data[data['Date'].isna()])
data
```

```python
from uuid import uuid4

data["eventid"] = [str(uuid4()) for _ in range(len(data))]
data["Year"] = pd.to_datetime(data["Date"], errors="coerce").dt.year.fillna(0).astype(int)
data["Month"] = pd.to_datetime(data["Date"], errors="coerce").dt.month.fillna(0).astype(int)
data["Day"] = pd.to_datetime(data["Date"], errors="coerce").dt.day.fillna(0).astype(int)

columns_with_defaults = {
    "City": "Unknown",
    "Country": "Unknown",
    "Perpetrator": "Unknown Group",
    "Weapon": "Unknown Weapon",
    "Injuries": 0,
    "Fatalities": 0,
    "Description": "No description available",
}

for col, default in columns_with_defaults.items():
    if col not in data.columns:
        data[col] = default
    data[col] = data[col].fillna(default)

data.rename(
    columns={
        "Perpetrator": "Group",
        "Weapon": "weapon_type",
        "Injuries": "Wounded",
        "Fatalities": "killed",
        "Description": "Summary",
        'City': 'city',
    },
    inplace=True,
)

columns_to_remove = [
    "City", "Perpetrator", "Weapon", "Injuries", "Fatalities", "Description", "date"
]
data.drop(columns=[col for col in columns_to_remove if col in data.columns], inplace=True)

data.reset_index(drop=True, inplace=True)
data

```

```python
merged_df = pd.concat([df, data], ignore_index=True)
merged_df.columns
```

```python
merged_df = merged_df.drop_duplicates()
```

```python
merged_df.to_csv('normalized_data.csv', index=False)
print('Successfully saved normalized data')
```
